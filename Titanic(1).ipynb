```python
# Titanic Dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

titanic = pd.read_excel("C:/Users/HP/Downloads/CTBA M1_Titanic.xlsx", sheet_name = "M1_Titanic")
print(titanic)
```

```python
#MWC Titanic Counts
mwc_count = titanic["MWC"].value_counts()
print(mwc_count)

# MWC Death Counts
dead_passengers = titanic[titanic["Survived"] == "Dead"]
dead_mwc_count = dead_passengers["MWC"].value_counts()
print(dead_mwc_count)

#MWC Survivor Counts
alive_passengers = titanic[titanic["Survived"] == "Alive"]
alive_mwc_count = alive_passengers["MWC"].value_counts()
print(alive_mwc_count)
```

```python
#Children by Sex Sums
children = titanic[titanic["MWC"] == "Child"]
children_sex_count = children["Sex"].value_counts()
print(children_sex_count)

#Children Death Sums
dead_children = children[children["Survived"] == "Dead"]
dead_children_sex_count = dead_children["Sex"].value_counts()
print(dead_children_sex_count)
```

```python
# Youngest Child
alive_children = children[children["Survived"] == "Alive"]
print(min(alive_children["Age"]))
youngest = alive_children[alive_children["Age"] == min(alive_children["Age"])]
print(youngest)
```

```python
# Millvina's family
dean_family = titanic[titanic['Name'].str.contains("DEAN")]
print(dean_family)
```

```python
mwc = np.array([1652, 432, 124])
mwclabels = ["Men", "Women", "Children"]
mwccolors = ['#377eb8', '#ff7f00', '#4daf4a']

plt.pie(mwc, labels = mwclabels, colors = mwccolors)
plt.title('Men, Women and Children Proportions')
plt.show()
```

```python
mwcsurvivors = np.array([321, 327, 64])
plt.pie(mwcsurvivors, labels = mwclabels, colors = mwccolors)
plt.title('Men, Women and Children Survivor Proportions')
plt.show()
```

```python
female_survivors = titanic[(titanic['Sex'] == 'Female') & (titanic['Survived'] == 'Alive')]
survfembyclass = female_survivors['Class'].value_counts()
print(survfembyclass)
```

```python
female_survivors = titanic[(titanic['Sex'] == 'Female') & (titanic['Survived'] == 'Alive')]
survfembyclass = female_survivors['Class'].value_counts()
survfembyclass.plot(kind='bar', color='skyblue')
plt.title('Female Survivors by Class')
plt.ylabel('Survivors')
plt.xlabel('Class')
```

```python
plt.hist(female_survivors['Age'], bins = 15)
plt.title('Female Survivor Age Distribution')
plt.ylabel('Survivor Count')
plt.xlabel('Age')
plt.show()
```
